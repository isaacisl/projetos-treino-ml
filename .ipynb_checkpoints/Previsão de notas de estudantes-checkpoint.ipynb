```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "bangladesh_student_performance.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "ihasan88/student-performance-dataset",
  file_path,
  # Provide any additional arguments like 
  # sql_query or pandas_kwargs. See the 
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)
```

```python
df = df.drop(columns=['Student_ID'], axis=1)
```

```python
X = df.drop("Previous_GPA", axis=1)
y = df.Previous_GPA
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

df.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer

cat_cols = ['Gender', 'District', 'School_Type', 'Parent_Education', 'Internet_Access', 'Private_Tuition' ]
num_cols = [col for col in X_train.columns if col not in cat_cols]


preprocessing = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
     ])

X_train_prep = preprocessing.fit_transform(X_train)
X_test_prep = preprocessing.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```
